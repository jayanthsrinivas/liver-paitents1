```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:/Users/jayanth/Downloads/Indian-liver-patients.csv")
data.info()
data.describe()
```

```python
data.loc[data["Age of the patient"]<=20,:].shape# below age 20 we have 41 patients
data.loc[data["Age of the patient"]<=20,:].head()
data.loc[data["Total Bilirubin"]>=20,:].shape# we have 19 people having bilirubin more than 20units, where the mean is 3.298
data.loc[data["Total Bilirubin"]>=20,:]
data.loc[data["Sgpt Alamine Aminotransferase"]>=1000,:]
data.loc[data["Sgot Aspartate Aminotransferase"]>=2000,:]
```

```python
#missing data
data_null=data[data.isnull().any(axis=1)]
data_null
```

```python
#Imputing through mean
data.isnull().sum()
mean1=round(data["Albumin and Globulin Ratio"].mean(),2)
data["Albumin and Globulin Ratio"]=data["Albumin and Globulin Ratio"].fillna(mean1)
data.loc[[209,241,253,312],:]
```

```python
x=data["Albumin"]/data["Albumin and Globulin Ratio"]
x=pd.DataFrame(x)
data=pd.concat([data,x],1)
data.rename(columns={0:"Globulin"},inplace=True)
data.head()
```

```python
#let us avoid few columns like Direct Bilirubin and Sgot
data_corr=data.iloc[:,[0,1,2,4,5,7,8,11,10]]
sns.heatmap(data_corr.corr())
data_corr.isnull().sum()
```

```python
male_female_patients=data.groupby(["Gender of the patient","Liver patient or not"]).agg({"Liver patient or not":"count"})# 1 being patient
#male_female_patients=pd.DataFrame(male_female_patients)
total_patients=data.groupby(["Gender of the patient"]).agg({"Gender of the patient":"count"})
male_female_patients
```

```python
for i in range(data_corr.shape[0]):
    if data_corr.iloc[i,8]==2:
        data_corr.iloc[i,8]=0
```

```python
data_corr_d=pd.get_dummies(data=data_corr)
data_corr_d.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x=data_corr_d.drop("Liver patient or not",1)
y=data_corr_d["Liver patient or not"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=60)
```

```python
x.isnull().sum()
y_test=pd.DataFrame(y_test)
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
y_pred = logreg.predict(x_test)
accuracy_score(y_test,y_pred)
```
